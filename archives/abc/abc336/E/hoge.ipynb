```python
def digit_sum(n:int):
    return sum(map(int, str(n)))
```

```python
for i in range(100):
    print(i+1, digit_sum(i+1))
```

```python
cnt = 0
for i in range(2024):
    cnt+= (i+1)% digit_sum(i+1)==0
    print(i+1, cnt)
```

```python
from collections import Counter
cnt = Counter()
for i in range(10**7):
    cnt[digit_sum(i+1)]+=1

print(cnt)
```

```python
import pandas as pd

df = pd.DataFrame(cnt.items(), columns=['digit_sum', 'count'])
df.plot(x='digit_sum', y='count', kind='bar', figsize=(10, 5))
```
